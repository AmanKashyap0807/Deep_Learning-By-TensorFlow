# Handwritten Digit Recognition using K-Nearest Neighbors (KNN)

## Introduction
Handwritten digit recognition is a classification problem where the goal is to classify images of handwritten digits (0-9) into their respective categories. The K-Nearest Neighbors (KNN) algorithm is a simple, yet effective supervised learning technique used for this purpose.

## Why K-Nearest Neighbors?
KNN is a non-parametric and lazy learning algorithm that makes predictions based on the majority class of its nearest neighbors in feature space. It is well-suited for handwritten digit recognition due to its simplicity and ability to handle multi-class classification problems.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import numpy as np
```

```python
X_train = np.load('X_train.npy')
X_test = np.load('X_test.npy')
y_train = np.load('y_train.npy')
y_test = np.load('y_test.npy')
```

```python
X_train.shape
```

## Flattening the images

```python
X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = knn.score(X_test, y_test)
print(f"test accuracy: {accuracy:.4f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('actual')
plt.title('Confusion Matrix')
plt.show()
```
